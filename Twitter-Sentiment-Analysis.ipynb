# Data Preprocessing

First of all we will split this data in training and test

```python
train,test= train_test_split(dataset,test_size=0.1)   #function 0.1 means we are taking 10% for testing 
```

We have to remove all the rt@.... , all hastags (#) - for this we have imported regix
->Now we will make a pattern that is we will develop a regix

```python
pattern = "(#\w+)|(RT\s@\w+:)|(http.*)|(@\w+)"  #regix is complete
```

Now , we will import constructor of Lemmatizer and Stemming

```python
#for stemming

ps = PorterStemmer()

#for lemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
def Clean_text(data):
    tweets = [] #array of tweets
    sentiments = [] #array of corrosponding Sentiments
    for index,row in data.iterrows():
        sentence = re.sub(pattern,'',row.text)    # ('pattern',to_replace,in what)
        words = [e.lower() for e in sentence.split()] # making all words lower case , we are splitting in space
        words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')] # lemmatization
        words = ' '.join(words)# now joining word in form of sentence
        tweets.append(words)
        sentiments.append(row.sentiment)
    return tweets,sentiments
        
```

Let's try this by passing training data

```python
train_tweets,train_sentiments = Clean_text(train)
```

```python
train_tweets  # we can stopwords , # , @ words are removed
```

Now lets make data_frame so that it becomeeasy for us to process

```python
final_data = {'tweets':train_tweets,'sentiments':train_sentiments} #This is our Dictionary
```

```python
#construct a data frame
processed_data = pd.DataFrame(final_data)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
processed_data['sentiments'] = labelencoder.fit_transform(processed_data['sentiments'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer #tfid can increase accuracy
cv= CountVectorizer(ngram_range=(1,3))      # If a sentence has 6 word it will make vector of maximum 3 words
cv.fit(processed_data['tweets'])        # As we have to fir count vectorin tweets colums
```
